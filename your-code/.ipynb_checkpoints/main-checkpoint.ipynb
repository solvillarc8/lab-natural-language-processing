## Now, we have to clean the html code removing words

- First we remove inline JavaScript/CSS
- Then we remove html comments. This has to be done before removing regular tags since comments can contain '>' characters
- Next we can remove the remaining tags

```python
# Your code
import re

def clean_html(text):
    # Remove inline JavaScript
    text = re.sub(r'<script.*?>.*?</script>', '', text, flags=re.DOTALL | re.IGNORECASE)

    # Remove inline CSS
    text = re.sub(r'<style.*?>.*?</style>', '', text, flags=re.DOTALL | re.IGNORECASE)

    # Remove HTML comments
    text = re.sub(r'<!--.*?-->', '', text, flags=re.DOTALL)

    # Remove remaining HTML tags
    text = re.sub(r'<.*?>', ' ', text)

    return text

data_train['clean_text'] = data_train['text'].apply(clean_html)
data_val['clean_text'] = data_val['text'].apply(clean_html)

data_train[['text', 'clean_text']].head()
```

- Remove all the special characters
    
- Remove numbers
    
- Remove all single characters
 
- Remove single characters from the start

- Substitute multiple spaces with single space

- Remove prefixed 'b'

- Convert to Lowercase

```python
# Your code
import re

def preprocess_text(text):

    # Remove HTML tags first (if needed)
    text = re.sub(r'<.*?>', ' ', text)

    # Remove all special characters
    text = re.sub(r'\W', ' ', str(text))

    # Remove all numbers
    text = re.sub(r'\d+', ' ', text)

    # Remove all single characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)

    # Remove single characters from the start
    text = re.sub(r'^\s*[a-zA-Z]\s+', ' ', text)

    # Remove prefixed 'b'
    text = re.sub(r'^b\s+', '', text)

    # Substitute multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)

    # Convert to lowercase
    text = text.lower()

    return text

data_train['processed_text'] = data_train['text'].apply(preprocess_text)
data_val['processed_text'] = data_val['text'].apply(preprocess_text)

data_train[['text', 'clean_text', 'processed_text']].head()
```

## Now let's work on removing stopwords
Remove the stopwords.

```python
# Your code
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))


def remove_stopwords(text):
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)

data_train['processed_text'] = data_train['processed_text'].apply(remove_stopwords)
data_val['processed_text'] = data_val['processed_text'].apply(remove_stopwords)
```

## Tame Your Text with Lemmatization
Break sentences into words, then use lemmatization to reduce them to their base form (e.g., "running" becomes "run"). See how this creates cleaner data for analysis!

```python
# Your code
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

data_train['token_processed_text'] = data_train['processed_text'].apply(word_tokenize)
data_val['token_processed_text'] = data_val['processed_text'].apply(word_tokenize)

def get_wordnet_pos(word):
    """Map the first letter of an NLTK POS tag to a WordNet constant."""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {'J': wordnet.ADJ, 'N': wordnet.NOUN,
                'V': wordnet.VERB, 'R': wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

lemmatizer = WordNetLemmatizer()

def lemmatize_row(tokens):
    """Lemmatize each token using its POS tag."""
    return [lemmatizer.lemmatize(w, get_wordnet_pos(w))
            for w in tokens]

data_train['Lemmatized'] = data_train['token_processed_text'].apply(lemmatize_row)
data_val['Lemmatized'] = data_val['token_processed_text'].apply(lemmatize_row)
```

## Bag Of Words
Let's get the 10 top words in ham and spam messages (**EXPLORATORY DATA ANALYSIS**)

```python
data_train.head()
```

```python
from collections import Counter
def get_top_words(token_lists, n=30):
    """Tokenize, lowercase, remove stopwords, return top-n word frequencies."""
    all_words = []
    for tokens in token_lists:
        all_words.extend([w for w in tokens if w not in stop_words and len(w) > 2])
    return Counter(all_words).most_common(n)

fake_titles = data_train.loc[data_train['label'] == 0, 'Lemmatized']
real_titles = data_train.loc[data_train['label'] == 1, 'Lemmatized']

top_fake = get_top_words(fake_titles)
top_real = get_top_words(real_titles)

# --- Bar charts ---
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
for ax, top, title, color in [
    (axes[0], top_fake[:20], 'Fake â Top 20 Title Words', '#f4a6a6'),
    (axes[1], top_real[:20], 'Real â Top 20 Title Words', '#a8ddb5'),
]:
    words, counts = zip(*top)
    ax.barh(words[::-1], counts[::-1], color=color, edgecolor='black')
    ax.set_title(title)
    ax.set_xlabel('Frequency')

plt.suptitle('Most Frequent Words in Titles', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Extra features

```python
# We add to the original dataframe two additional indicators (money symbols and suspicious words).
money_simbol_list = "|".join(["euro","dollar","pound","â¬",r"\$"])
suspicious_words = "|".join(["free","cheap","sex","money","account","bank","fund","transfer","transaction","win","deposit","password"])

data_train['money_mark'] = data_train['token_processed_text'].str.contains(money_simbol_list)*1
data_train['suspicious_words'] = data_train['token_processed_text'].str.contains(suspicious_words)*1
data_train['text_len'] = data_train['token_processed_text'].apply(lambda x: len(x)) 

data_val['money_mark'] = data_val['token_processed_text'].str.contains(money_simbol_list)*1
data_val['suspicious_words'] = data_val['token_processed_text'].str.contains(suspicious_words)*1
data_val['text_len'] = data_val['token_processed_text'].apply(lambda x: len(x)) 

data_train.head()
```

## How would work the Bag of Words with Count Vectorizer concept?

```python
# Your code
from sklearn.feature_extraction.text import CountVectorizer

data_train['lemmatized_text'] = data_train['Lemmatized'].apply(lambda x: " ".join(x))
data_val['lemmatized_text'] = data_val['Lemmatized'].apply(lambda x: " ".join(x))

bow_vectorizer = CountVectorizer(
    max_features=5000,
    min_df=2,
    max_df=0.95
)

# Fit on train, transform both
X_train_bow = bow_vectorizer.fit_transform(data_train['lemmatized_text'])
X_val_bow = bow_vectorizer.transform(data_val['lemmatized_text'])

# Shapes
print("Train shape:", X_train_bow.shape)
print("Validation shape:", X_val_bow.shape)
```
